```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
link='https://en.wikipedia.org/wiki/List_of_the_largest_software_companies'
res=requests.get(link)
```

```python
soup=BeautifulSoup(res.text,'html.parser')
```

```python
table = soup.find("table", class_="wikitable sortable")

header = []

for h in table.find_all('th'):
    header.append(h.text.strip())
    
rows=[]
for row in table.find_all('tr')[1:]:
    data=row.find_all('td')
    datas=[]
    for d in data:
        datas.append(d.text.strip())
    rows.append(datas)
```

```python
df = pd.DataFrame(rows, columns=header)
df.to_csv("CompaniesData.csv",index=False)
```
