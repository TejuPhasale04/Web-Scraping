```python
import chromedriver_binary
from selenium.webdriver.chrome.service import Service
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
service=Service(chromedriver_binary.chromedriver_filename)
```

```python
driver=webdriver.Chrome(service=service)
link="https://stockmages.netlify.app/"
```

```python
driver.get(link)
```

```python
soup=BeautifulSoup(driver.page_source,'html.parser')
```

```python
data = []
for sp in soup.find_all('div', class_='container'):
    # Extracting the image link
    link = sp.find('img').get('src')
    print(link)

    # Extracting the tag text
    tag = sp.find('span', class_='tag-color').text
    tag = tag.replace('Tags -', '').strip()
    tags=tag.split(',')
    print(list(set(tags)))

    # Extracting likes and comments
    likeComments = []
    i = sp.find('div', class_='likes-comments')
    for likeComment in i.find_all('span'):
        likeComments.append(likeComment.text.strip())

    # Clean up the likes and comments
    likes = likeComments[0].replace('Likes', '').strip() if len(likeComments) > 0 else ''
    comments = likeComments[1].replace('Comments', '').strip() if len(likeComments) > 1 else ''
    print([likes, comments])

    # Append the extracted data
    data.append([tag, link, likes, comments])

# Output the data list
print(data)
```

```python
df=pd.DataFrame(data,columns=['Tags','Links','Likes','Comments'])
```

```python
df.to_csv('Image Data.csv',index=False)
```

### Download Multiple Images

```python
path=[]
for link in df['Links']:
    res=requests.get(link)
    Img=link.split('/')[-1]
    fd=open('Img/'+Img,'wb')
    fd.write(res.content)
    fd.close()
    path.append('Img/'+Img)
```

```python
df['Path']=path
```

```python
df.to_csv('Image Data.csv',index=False)
```
